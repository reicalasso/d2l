# 1.4 Roots

Roots of Machine Learning:
Machine learning builds on centuries of work in mathematics, statistics, and science. Early foundations include probability distributions (Bernoulli, Gaussian) and optimization methods like least squares (Gauss), which enabled modeling real-world phenomena with data. Early statistical ideas such as averaging and trimmed means appeared as far back as the Middle Ages.

Modern statistics advanced with Ronald Fisher, whose methods (e.g., linear discriminant analysis, Fisher information) and datasets (Iris) remain influential, while also reminding us that data science can be misused ethically. Additional roots come from information theory (Claude Shannon) and computation theory (Alan Turing), including the Turing Test for machine intelligence.

Neuroscience and psychology inspired learning algorithms, notably Hebbian learning (“neurons that fire together wire together”), leading to perceptrons and stochastic gradient descent. Neural networks are based on layered linear and nonlinear transformations trained via backpropagation.

Despite early promise, neural networks declined (1995–2005) due to limited computation and small datasets. During this period, methods like kernel techniques, decision trees, and graphical models dominated because they were faster to train and had stronger theoretical guarantees.
